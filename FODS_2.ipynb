```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("bitcoin.csv")
```

Display the first 4 rows

```python
print("First 4 rows of the dataset:")
print(df.head(4))
print("\n" + "="*50 + "\n")
```

Display the features in the dataset

```python
print("Features in the dataset:")
print(df.columns)
print("\n" + "="*50 + "\n")
```

Drop duplicates and print the number of duplicates dropped

```python
initial_shape = df.shape
df.drop_duplicates(inplace=True)
print(f"Number of duplicates dropped: {initial_shape[0] - df.shape[0]}")
print("\n" + "="*50 + "\n")
```

Fill missing values with the previous value in the column (forward fill)

```python
df.fillna(method='ffill', inplace=True)
print("Missing values after forward fill:")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")
```

Sort the dataset by 'Date'

```python
df.sort_values(by='Date', inplace=True)
print("Dataset sorted by date:")
print(df.head())
print("\n" + "="*50 + "\n")
```

Add a new column 'Daily Change' to represent the change in price

```python
df['Daily Change'] = df['Close'] - df['Open']
print("First 4 rows with Daily Change:")
print(df.head(4))
print("\n" + "="*50 + "\n")
```

Get the rolling average of the 'Close' price for 7 days

```python
df['7-Day Rolling Avg'] = df['Close'].rolling(window=7).mean()
print("First 4 rows with 7-Day Rolling Avg:")
print(df.head(4))
print("\n" + "="*50 + "\n")
```

Calculate the percentage change in closing price

```python
df['Pct Change'] = df['Close'].pct_change() * 100
print("First 4 rows with Percentage Change:")
print(df.head(4))
print("\n" + "="*50 + "\n")
```

Group by year and calculate the average closing price for each year

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
yearly_avg = df.groupby('Year')['Close'].mean()
print("Yearly average closing price:")
print(yearly_avg)
print("\n" + "="*50 + "\n")
```

Get the index of the row with the maximum closing price

```python
max_price_idx = df['Close'].idxmax()
print(f"Row index with the maximum closing price: {max_price_idx}")
print("Data for the day with maximum closing price:")
print(df.loc[max_price_idx])
print("\n" + "="*50 + "\n")
```

Filter the dataset for a specific year

```python
df_2017 = df[df['Year'] == 2017]
print("Data for the year 2017:")
print(df_2017.head())
print("\n" + "="*50 + "\n")
```

Line chart


```python
 plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='blue')
plt.plot(df['Date'], df['7-Day Rolling Avg'], label='7-Day Rolling Avg', color='orange')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bitcoin Closing Price and 7-Day Rolling Average Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Bar Plot

```python
plt.figure(figsize=(10, 6))
yearly_avg.plot(kind='bar', color='purple')
plt.xlabel('Year')
plt.ylabel('Average Closing Price')
plt.title('Yearly Average Closing Price of Bitcoin')
plt.show()
```

Pie chart

```python
price_increase_count = df['Daily Change'].apply(lambda x: 'Increase' if x > 0 else 'Decrease').value_counts()
plt.figure(figsize=(8, 8))
plt.pie(price_increase_count, labels=price_increase_count.index, autopct='%1.1f%%', startangle=140, colors=['green', 'red'])
plt.title('Distribution of Price Increases vs. Decreases')
plt.show()
```

KDE plot

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Close'], shade=True, color='blue')
plt.xlabel('Price')
plt.title('Density Plot of Bitcoin Closing Prices')
plt.show()
```

Line plot

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Pct Change'], label='Percentage Change', color='green')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.title('Bitcoin Percentage Change Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
